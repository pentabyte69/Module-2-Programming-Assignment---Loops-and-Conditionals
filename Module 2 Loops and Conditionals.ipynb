```python
# Assigning the numbers 
secret = 7 
guess = 5
print (f"Secret: {secret}, Guess: {guess}")
# Conditional tests
if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

```python
# Possible items:
# cherry: small, not green
# pea: small, green
# watermelon: not small, green
# pumpkin: not small, not green
# Test 1
small = True
green = True
print(f"Test 1: Small={small}, Green={green}")
if small and green:
      print("This matches: pea")
elif small and not green:
      print ("This matches: cherry")
elif not small and green:
      print("This matches: watermelon")
else: # not small and not green
      print("This matches: pumpkin")
```

```python
# We want the sequence: 3,2,1,0
# - start: 3 (inclusive)
# - stop: -1 (exclusive, so it stops *after* 0)
# - step: -1 (to count down by 1)

print(list(range(3,-1,-1)))
```

```python
guess_me = 7
number = 1

while True:
    print(f"Checking number: {number}")
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it')
        break  # Exit the loop
    else: # number > guess_me
        print('oops')
        break  # Exit the loop
        
    # Increment number at the end of the loop
    number += 1
```

```python
guess_me = 5

for number in range(10):
    print(f"Checking number: {number}")
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break  # Exit the for loop
    else: # number > guess_me
        print('oops')
        break # This break also
```
